```python
"""
projekt_1.py: první projekt do Engeto Online Python Akademie
author: David Slavík
email: slavikdavid4@gmail.com
discord: _africa
"""
TEXTS = ['''
Situated about 10 miles west of Kemmerer,
Fossil Butte is a ruggedly impressive
topographic feature that rises sharply
some 1000 feet above Twin Creek Valley
to an elevation of more than 7500 feet
above sea level. The butte is located just
north of US 30N and the Union Pacific Railroad,
which traverse the valley. ''',
'''At the base of Fossil Butte are the bright
red, purple, yellow and gray beds of the Wasatch
Formation. Eroded portions of these horizontal
beds slope gradually upward from the valley floor
and steepen abruptly. Overlying them and extending
to the top of the butte are the much steeper
buff-to-white beds of the Green River Formation,
which are about 300 feet thick.''',
'''The monument contains 8198 acres and protects
a portion of the largest deposit of freshwater fish
fossils in the world. The richest fossil fish deposits
are found in multiple limestone layers, which lie some
100 feet below the top of the butte. The fossils
represent several varieties of perch, as well as
other freshwater genera and herring similar to those
in modern oceans. Other fish such as paddlefish,
garpike and stingray are also present.'''
]

accounts = {
    "bob": "123",
    "ann": "pass123",
    "mike": "password123",
    "lizz": "pass123"
    }

# Uživatel zadává přihlašovací jméno
(username := input("Enter username:"))
print(f"username: {username}")
# Uživatel zadává přihlašovací heslo
if (password := input("Enter password:")) == accounts.get(username):
    print(f"password: {password}")
    print("-" * 40)
    print(f"Welcome to the app, {username}\nWe have 3 texts to be analyzed.")
    # Uživatel zadává číslicí který text chce analyzovat (1-3)
    choice = input("Enter a number btw. 1 and 3 to select:")
    if choice.isdigit():
        choice = int(choice)
        if 1 <= choice <= len(TEXTS):
            print("-" * 40)
            print(f"Enter a number btw. 1 and 3 to select: {choice}")
            print("-" * 40)
            words = TEXTS[choice -1].rsplit()
            cleaned_words = [word.rstrip(",.") for word in words]
            upper_words = 0 # Pomocná proměnná pro slova začínající velkým písmenem
            upper_all_words = 0 # Pro slova velkými písmeny
            lower_word = 0  # Pro slova malými písmeny
            numbers = 0 # Pro počet čísel (ne cifer)
            numbers_sum = 0 # Pro součet čísel  
            word_length= {} # Pro délku slov 
            # Iterace očištěných slov od teček a čárek a následné sčítání
            for cleaned_word in cleaned_words:
                length = len(cleaned_word)
                if length in word_length:
                    word_length[length] += 1
                else:
                    word_length[length] = 1
                if cleaned_word[0].isupper():
                    upper_words += 1
                elif cleaned_word.isupper():
                    upper_all_words += 1
                elif cleaned_word.islower():
                    lower_word += 1
                elif cleaned_word.isdigit():
                    numbers += 1
                    numbers_sum += int(cleaned_word)
            print(f"There are {len(cleaned_words)} words in the selected text.")
            print(f"There are {upper_words} titlecase words.")
            print(f"There are {upper_all_words} uppercase words.")
            print(f"There are {lower_word} lowercase words.")
            print(f"There are {numbers} numeric strings.")
            print(f"The sum of all the numbers is {numbers_sum}.")
            print("-" * 40)
            # Záhlaví tabulky, zarovnání dle největší hodnoty ze slovníku a přičtení 2 pro lepší přehled
            print("LEN|{:^{width}}|NR.".format('OCCURENCES', width=max(word_length.values())+2))
            print("-" * 40)
            # Iterace seřazených klíčů a hodnot pro zobrazení tabulky
            for key in (sorted(word_length.keys())):
                char = "*"  # Pomocná proměnná pro výpis znaku
                count = word_length.get(key)    # Pro zarovnání dle počtu hodnot a přičtení 2 pro lepší přehled
                print(f"{key:>3}|{char * count:<{max(word_length.values())+2}}|{count}")         
        else:
            print("Invalid choice. The number must be between 1 and 3.")
    else:
        print("Invalid input. Next time please enter a valid number.")
else:
    print(f"password: {password}")
    print("unregistered user, terminating the program..")
```
